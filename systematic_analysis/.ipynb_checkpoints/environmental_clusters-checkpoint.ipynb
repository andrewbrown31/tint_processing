```python
import pandas as pd
import numpy as np
import glob
import datetime as dt
import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def last_day_of_month(any_day):
    # get close to the end of the month for any day, and add 4 days 'over'
    next_month = any_day.replace(day=28) + dt.timedelta(days=4)
    # subtract the number of remaining 'overage' days to get last day of current month, or said programattically said, the previous day of the first of next month
    return next_month - dt.timedelta(days=next_month.day)    

def load_scws(rid,year,month):
    
    d1 = dt.datetime(year,month,1)
    d2 = last_day_of_month(d1)
    temp_df = pd.read_pickle("/g/data/eg3/ab4502/ExtremeWind/points/era5_aws_tint_"+rid+"_"+d1.strftime("%Y%m%d")+"_"+d2.strftime("%Y%m%d")+"_max.pkl")
    scws = temp_df.query("(gust>=25) & (in10km ==1)").sort_values("gust",ascending=False)
    scws["day"] = pd.DatetimeIndex(scws["dt_utc"]).floor("1D")
    scws = scws.drop_duplicates(subset=["day"])
    scws = scws.set_index(pd.DatetimeIndex(scws.dt_utc)).sort_index()
    
    return scws
```

```python
woom = pd.DataFrame()
for y in tqdm.tqdm(np.arange(2002,2021)):
    for m in np.arange(1,13):
        try:
            woom = pd.concat([woom, load_scws("27",y,m)])
        except:
            print("CAN'T LOAD YEAR "+str(y)+" MONTH "+str(m))
```

```python
namoi = pd.DataFrame()
for y in tqdm.tqdm(np.arange(2010,2021)):
    for m in np.arange(1,13):
        try:
            namoi = pd.concat([namoi, load_scws("69",y,m)])
        except:
            print("CAN'T LOAD YEAR "+str(y)+" MONTH "+str(m))
```

```python
melb = pd.DataFrame()
for y in tqdm.tqdm(np.arange(2008,2021)):
    for m in np.arange(1,13):
        try:
            melb = pd.concat([melb, load_scws("2",y,m)])
        except:
            print("CAN'T LOAD YEAR "+str(y)+" MONTH "+str(m))
```

```python
bris = pd.DataFrame()
for y in tqdm.tqdm(np.arange(2007,2021)):
    for m in np.arange(1,13):
        try:
            bris = pd.concat([bris, load_scws("66",y,m)])
        except:
            print("CAN'T LOAD YEAR "+str(y)+" MONTH "+str(m))
```

```python
syd = pd.DataFrame()
for y in tqdm.tqdm(np.arange(2010,2021)):
    for m in np.arange(1,13):
        try:
            syd = pd.concat([syd, load_scws("71",y,m)])
        except:
            print("CAN'T LOAD YEAR "+str(y)+" MONTH "+str(m))

#Drop the Kurnell tornado which was measured by Kurnell AWS
syd = syd.drop("2015-12-15 23:33:00")
```

```python
#Map all the clusters to how they are referred to in our first paper
melb["cluster_new"] = melb.cluster.map({1:2,2:1,0:0})
syd["cluster_new"] = syd.cluster.map({1:2,2:1,0:0})
bris["cluster_new"] = bris.cluster.map({1:2,2:1,0:0})
namoi["cluster_new"] = namoi.cluster.map({1:2,2:1,0:0})
woom["cluster_new"] = woom.cluster.map({1:2,2:1,0:0})

print(melb.cluster_new.value_counts())
print(syd.cluster_new.value_counts())
print(bris.cluster_new.value_counts())
print(namoi.cluster_new.value_counts())
print(woom.cluster_new.value_counts())
```

```python
data = pd.concat([melb, syd, bris], axis=0)
```

```python
#Linear
data.loc[(data.aspect_ratio>=3) & (data.major_axis_length>=100),"class2"] = "Linear"
#Non-linear
data.loc[(data.aspect_ratio<3) & (data.major_axis_length>=100),"class2"] = "Non-linear"
#Cellular
data.loc[(data.local_max == 1),"class2"] = "Cellular"
#Macroburst
data.loc[(data.local_max>=2) & (data.major_axis_length<100),"class2"] = "Cell cluster"
#Supercell
data.loc[(data.azi_shear>4.5) & ((data.aspect_ratio<3) | (data.major_axis_length<100)),"class2"] = "Supercellular"
#Shallow
#data.loc[(data.max_alt<8) & (data.conv_pct<.4),"class2"] = "Shallow"
```
