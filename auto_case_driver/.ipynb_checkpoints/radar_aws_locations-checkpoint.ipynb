```python
import pyart
import matplotlib.pyplot as plt
import numpy as np
import math
import warnings
import xarray as xr
import itertools
import pandas as pd
import datetime as dt
import glob as glob
import zipfile
warnings.simplefilter("ignore")
```

```python
def load_radar(rid, time, radar_t_delta=0):

    #time: Time of interest (in this context, the time of maximum BARPAC gust). datetime object
    #rid: Radar ID (https://www.openradar.io/operational-network). two-digit string
    #radar_t_delta: Add some interval (in minutes) to look at a time relative to the time of maximum model gust. integer
    
    time = time + dt.timedelta(minutes=radar_t_delta)
    
    isfile = len(glob.glob("/g/data/eg3/ab4502/radar/"+rid+"_"+time.strftime("%Y%m%d")+"*")) > 0
    
    if not isfile:
        print("INFO: UNPACKING RADAR FILE FOR ID "+rid+" AND TIME "+time.strftime("%Y%m%d"))
        path_to_zip_file = "/g/data/rq0/level_1/odim_pvol/"+rid+"/"+str(time.year)+"/vol/"+rid+"_"+time.strftime("%Y%m%d")+".pvol.zip"
        directory_to_extract_to = "/g/data/eg3/ab4502/radar/"
        with zipfile.ZipFile(path_to_zip_file, 'r') as zip_ref:
            zip_ref.extractall(directory_to_extract_to)
        
    else:
        print("INFO: LOADING RADAR FILE FOR ID "+rid+" AND TIME "+time.strftime("%Y%m%d"))
        
    files = glob.glob("/g/data/eg3/ab4502/radar/"+rid+"_"+time.strftime("%Y%m%d")+"*")
    f_times = [fname.split("/")[-1].split("_")[2].split(".")[0] for fname in files]
    radar_file=pyart.aux_io.read_odim_h5(\
                              files[np.argmin(np.array([abs(time-dt.datetime(time.year,time.month,time.day,int(f[0:2]),int(f[2:4]),int(f[4:6]))) for f in f_times]))],\
                              file_field_names=True)
    return radar_file
```

```python
min_lon=140;max_lon=150;min_lat=-40;max_lat=-33.5
radar02 = load_radar("02", time=dt.datetime(2020,1,31))
radar14 = load_radar("14", time=dt.datetime(2020,1,31))
radar30 = load_radar("30", time=dt.datetime(2020,1,31))
radar49 = load_radar("49", time=dt.datetime(2020,1,31))
radar68 = load_radar("68", time=dt.datetime(2020,1,31))
radar95 = load_radar("95", time=dt.datetime(2021,1,31))

barra = xr.open_dataset("/g/data/ma05/BARRA_R/v1/static/topog-an-slv-PT0H-BARRA_R-v1.nc").\
        sel({"longitude":slice(min_lon-1,max_lon+1),"latitude":slice(min_lat-1,max_lat+1)})
lons,lats = np.meshgrid(barra.longitude.values, barra.latitude.values)

#http://www.bom.gov.au/climate/data/lists_by_element/alphaVIC_284.txt
aws = pd.read_fwf("/home/548/ab4502/working/observations/vic_onemin_gust_sites.txt", header=None, 
                  names=["site","name","lat","lon","start1","start2","end1","end2","years","%","obs","AWS"])

plt.figure(figsize=[10,8])
ax = plt.gca()

def plot_radar(radar,embel=False,dop=True):
    disp=pyart.graph.RadarMapDisplayBasemap(radar)
    disp.plot_ppi_map("DBZH",1,cmap=pyart.graph.cm_colorblind.HomeyerRainbow, resolution="h",colorbar_label="DBHZ",\
        projection="merc",vmin=-100, vmax=-99,ax=ax, mask_outside=True,colorbar_flag=False,\
        min_lon=min_lon,max_lon=max_lon,min_lat=min_lat,max_lat=max_lat,embelish=embel)
    plt.title("")
    disp.basemap.plot([radar.longitude["data"][0]],[radar.latitude["data"][0]], marker="+", ms=15, mew=2, color="tab:orange", latlon=True, mfc="none")
    if dop:
        disp.plot_range_ring(150)
    else:
        disp.plot_range_ring(150,ls="--")
    return disp

disp=plot_radar(radar02,embel=True)
_=plot_radar(radar14,dop=False)
_=plot_radar(radar30,dop=False)
_=plot_radar(radar49)
_=plot_radar(radar68,dop=False)

disp.basemap.drawcoastlines(linewidth=2)
c=disp.basemap.contourf(lons,lats,barra["topog"].values,latlon=True,\
                        cmap=plt.get_cmap("terrain"),levels=[-1,0,150,300,450,600,750,900,1200],extend="both")
cb=plt.colorbar(c)

scat=disp.basemap.scatter(x=aws["lon"].values,y=aws["lat"].values,marker="^",s=75,latlon=True,c=aws["years"],cmap=plt.get_cmap("Reds"))
cb2=plt.colorbar(scat,cax=plt.axes([0.02,0.2,0.02,0.6]), orientation="vertical")

cb.ax.tick_params(labelsize='large')
cb.set_label("Terrain height (m)",size=16)

cb2.ax.tick_params(labelsize='large')
cb2.set_label("AWS observations (years)",size=16)

plt.savefig("/home/548/ab4502/working/observations/vic_aws_locs.png", quality=95, bbox_inches="tight")
```
