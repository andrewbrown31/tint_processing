```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/g/data/eg3/ab4502/figs/ExtremeWind/case_studies/storm_stats.csv")
cluster=pd.read_csv("/g/data/eg3/ab4502/figs/ExtremeWind/case_studies/cluster_case_output_era5.csv")
data["cluster"] = cluster.cluster.values #NOTE this assumes that the list of events in this notebook is the same as in kmeans_and_cluster_eval.ipynb
data["cluster"] = data.cluster.map({1:1,2:2,0:3})

round_dict = {"Major axis (km)":1, "Aspect ratio":1, "Aziumthal shear (s^-1)":4, "Convective pixel fraction":2}

data["Aziumthal shear (s^-1)"] = data["Aziumthal shear (s^-1)"]/1000
data = data.drop(columns=["azi_shear2","Convective pixel fraction"])\
    .rename(columns={"cluster":"Environ. cluster","time":"Time (UTC)","Location":"Location (Figure reference)"}).round(
    round_dict).replace(np.nan,"")#.set_index("Time (UTC)")


data_rearranged = data.loc[list(np.arange(0,18)) + [18,19,22,23,20,24,25,21,26] + list(np.arange(27,36))]
data_rearranged["Location (Figure reference)"] = data["Location (Figure reference)"]
data_rearranged.loc[22,"Location (Figure reference)"] = "Oakey (6c)"
data_rearranged.loc[23,"Location (Figure reference)"] = "Oakey (6d)"
data_rearranged.loc[20,"Location (Figure reference)"] = "Amberley (6e)"
data_rearranged.loc[24,"Location (Figure reference)"] = "Oakey (6f)"
data_rearranged.loc[25,"Location (Figure reference)"] = "Oakey (6g)"
data_rearranged.loc[21,"Location (Figure reference)"] = "Amberley (6h)"
data_rearranged.loc[26,"Location (Figure reference)"] = "Oakey (6i)"

data_rearranged.to_latex("/g/data/eg3/ab4502/figs/ExtremeWind/case_studies/storm_stats.tex",index=False)
```
